```python
import pandas as pd
from utils import *

#Paises con mayores costos de importacion 
path = get_arch();

df = pd.read_excel(path,nrows=55,engine='openpyxl')
columna_valores_de_importacion = df['Valor_CIF_USD']
paises = df['País_origen']

df_grouped = df.groupby('País_origen')['Valor_CIF_USD'].mean()

# Ordena los datos en orden descendente
df_sorted = df_grouped.sort_values(ascending=False)

# Convierte la serie en un DataFrame y reinicia el índice
df_final = df_sorted.reset_index()

# Muestra los países con los mayores precios de importación
df_final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de ejemplo
x = df_final['País_origen']
y = df_final['Valor_CIF_USD']

# Crea una figura y un conjunto de ejes
fig, ax = plt.subplots(figsize=(10, 8))

# Personaliza el fondo de la figura
fig.patch.set_facecolor('#ffffff')

# Define una paleta de colores
palette = sns.color_palette("rocket", as_cmap=True)

# Dibujar los datos en el conjunto de ejes
ax.plot(x, y, color=palette(0.7), linewidth=3, marker='o', markerfacecolor=palette(0.9), markeredgecolor=palette(0.7))

# Personalizar el gráfico
ax.set_title('Promedio de Importación por País', fontsize=16, fontweight='bold', fontfamily='Arial')
ax.set_xlabel('País de Origen', fontsize=14, fontfamily='Arial')
ax.set_ylabel('Promedio de Valor CIF (USD)', fontsize=14, fontfamily='Arial')

# Personaliza las etiquetas de los ejes
plt.xticks(rotation=45, fontsize=12, fontfamily='Arial')
plt.yticks(fontsize=12, fontfamily='Arial')

# Agrega una cuadrícula al gráfico
ax.grid(color='#cfcfcf', linestyle='-', linewidth=1)

# Agrega una leyenda
ax.legend(loc='best', fontsize=12, frameon=False)

# Muestra el gráfico
plt.show()
```
