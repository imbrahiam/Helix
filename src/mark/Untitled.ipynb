```python
import matplotlib.pyplot as plt
import common
import pandas as pd
import drawmap

ruta = '../raw_data/Perfil_EmpresasImpo_2022_WEB.xlsx'

columnas = ['Region_Ubicacion', 'Valor_CIF_USD']

df = pd.read_excel(ruta, usecols=columnas)
df
```

```python
df_grouped = df.groupby(columnas[0])[columnas[1]].sum().reset_index()
df_grouped
```

```python
df_grouped_sorted = df_grouped.sort_values(columnas[1])
df_grouped_sorted = df_grouped_sorted.reset_index().drop('index', axis=1)
df_grouped_sorted = df_grouped_sorted.set_index(columnas[0])
df_grouped_sorted

draw_hplot('Costo', 'Region', df_grouped_sorted, columnas[1])

df_grouped_sorted
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

df_grouped_sorted = df_grouped_sorted.sort_values(columnas[1], ascending = False)
x = df_grouped_sorted.index
y = df_grouped_sorted[columnas[1]]

plt.bar(x, y, width=0.6)

plt.xticks(rotation=90)

plt.ylabel('Costo')
plt.xlabel('Region')
plt.title('Costo de importacion por region (Ordenado)')

plt.tight_layout()  # Ajusta automáticamente los parámetros del subplot para que el gráfico se ajuste al área de dibujo
plt.show()
```
